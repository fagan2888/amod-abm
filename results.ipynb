```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt

NT = [
    [380,380,380],
    [571,571,571],
    [827,827,827],
    [1146,1146,1146]
]

FS = [
    [40,60,80],
    [60,80,100],
    [80,100,120],
    [100,120,140]
]

SR = [
    [92.2,96.8,99.0],
    [93.0,95.6,96.8],
    [93.1,97.1,97.9],
    [93.9,94.7,96.8]
]

VMT = [
    [94.0,71.2,57.5],
    [86.7,73.9,61.6],
    [84.2,73.4,66.5],
    [85.5,78.2,69.4]
]

WT = [
    [239.8,173.2,136.2],
    [199.4,162.4,131.1],
    [183.1,161.3,135.5],
    [177.8,155.7,136.0]
]

VT = [
    [478.5,447.1,471.1],
    [469.8,443.0,442.5],
    [458.8,460.7,455.2],
    [456.2,452.8,464.7]
]

TT = np.zeros((4,3))
for i in range(4):
    for j in range(3):
        TT[i][j] = 1.5*WT[i][j] + VT[i][j]
TVMT = np.zeros((4,3))     
for i in range(4):
    for j in range(3):
        TVMT[i][j] = VMT[i][j] * FS[i][j]

Colors = [
    [0,0,0],
    [1,1,1],
    [2,2,2],
    [3,3,3]
]
```

```python
plt.figure()
handles = []
for i in [3,2,1,0]:
    handle = plt.scatter(FS[i], TT[i], s=VMT[i],
                         c='r' if i == 0 else 'y' if i == 1 else 'g' if i == 2 else 'b',
                         alpha=0.5)
    handles.append(handle)
plt.legend(handles, ("MCM-3.5", "MCM-4.0", "MCM-4.5", "MCM-5.0"), scatterpoints=1)
plt.ylabel('Generalized Travel Time (s)')
plt.xlabel('Fleet Size')
plt.ylim((500,900))
plt.show()
```
